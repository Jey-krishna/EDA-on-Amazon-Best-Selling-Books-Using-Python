### Import necessary packages

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

### Explore and understand the data

```python
df = pd.read_csv("bestsellers with categories.csv")
df
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

## Data Visualization 

```python
sns.countplot(data=df, x= 'User Rating')
```

```python
plt.title('Genre vs Review')
sns.barplot(df, x= 'Genre', y = 'Reviews')
plt.show()
```

## TOP 10 Analysis

```python
top_Authers = df.sort_values('Price', ascending=False)[['Name','Author','User Rating', 'Reviews', 'Price']].head(10)

top_Authers
```

### Top Authors by user rating

```python

plt.figure(figsize=(15,3))
plt.title('Top Authors by user rating')
sns.barplot(top_Authers, x= 'Author', y = 'User Rating')
plt.show()
```

### Top books by price

```python
plt.figure(figsize=(5,3))
plt.title('Top books by Price')
sns.barplot(top_Authers, x= 'Price', y = 'Name')
plt.show()
```

### Top  books by user rating


```python
plt.figure(figsize=(5,3))
plt.title('Top books by Reviews')
sns.barplot(top_Authers, x= 'Reviews', y = 'Name')
plt.show()
```

### Top books by ratings


```python
plt.figure(figsize=(5,3))
plt.title('Top books by Reviews')
sns.barplot(top_Authers, x= 'User Rating', y = 'Name', )
plt.show()
```

### Distribution Visualization


Distributin of price of books

```python
plt.figure(figsize=(5,3))
plt.title('Distributin of price of books')
sns.histplot(data=df, x='Price',  hue= 'Genre', element='poly')
plt.show()

```

Distributin of User rating of books

```python
plt.figure(figsize=(5,3))
plt.title('Distributin of User rating of books')
sns.histplot(data=df, x='User Rating',  hue= 'Genre', element='poly')
plt.show()
```

Distributin of User rating of books

```python
plt.figure(figsize=(5,3))
plt.title('Distributin of User rating of books')
sns.histplot(data=df, x='Reviews', hue= 'Genre', element='poly')
plt.show()
```
